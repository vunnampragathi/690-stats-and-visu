it is difficult to analyse all the data.So i am analysing the data based on particular year.here i am choosing 2010 data for my analysis

```python
df1=df[df['Year']==2010]
df1
```

Even after selecting one year data there are total 185 rows, it is also diffult for analyses so i am choosing some popular countires like china,india,united staes,brazil,pakistan,russia for my further analyses.

```python
df2=df1[df1['Country Name'].isin(['Brazil','China','India','Russian Federation','Pakistan','United States'])]
df2
```

```python
# plotting the bar graph for forest area
px.bar(df2, x = 'Country Name', y = 'Forest area (% of land area)',color='Country Name',title='Bar chart showing the total forest area in selected countries')
```

from the above bar chart it is obsereved that brazil has highest forest area,it has 60% of total land as forests.russia is in the second place next is united staets, india china and last is pakistan.whereas pakistan has very forest area less than 10% of total land.from the above chart we can say that brazil has good atmospehric conditions compared to the other countries.

```python
# plotting the pie chart for green house gas emissions
px.pie(df2,values='Total greenhouse gas emissions (kt of CO2 equivalent)',color='Country Name',names='Country Name',title='pie chart showing the total green house gases emissions')
```

from the above pie chart it is obserevd that china has higest green house gas emissions it has approximately about 50% compared to other countries.next highest green house gas emissions is in united states,next is india and pakistan has very less amount of green house gas emissions.from this we can assume that china is the more pollutant country and pakistan is less pollutant compared to other countries.

i want to analyse how the air pollution is changing over the years for the selected countries.so i am selecting the countries from the original dataset

```python
df3=df[df['Country Name'].isin(['Brazil','China','India','Russian Federation','Pakistan','United States'])]
df3
```

```python
df3.shape
```

```python
# plotting the line chart of air pollution for selected countries from 2010 to 2017
px.line(df3,x='Year',y='PM2.5 air pollution, mean annual exposure (micrograms per cubic meter)',color='Country Name',title='line chart showing annual exposure of air pollution from 2010 to 2017')
```

from the above line chart it is observed that india is in first place in terms of polluted air,but compared to 2010 the pollution level is reduced in the 2017.china is in second place intially but it pollution levels are decreased as the years passed it is in the 3 rd place in 2017.it seems that USA has very fresh air.

```python
 # plotting the scatter plot for forest area and grrn house gas
 px.scatter(df1,
    x="Total greenhouse gas emissions (kt of CO2 equivalent)",
    y="Forest area (% of land area)", 
    color='Country Name',
    title="scatter plot showing total forest VS total green house gas emissions across all the countries")
```

from the above scatter plot it is observed that china has highest green house gases emissions compared to all other countries with forest area approximately about 20% of total land.next place is for usa.it seems that as the forest area reduces there is chance of more emmission of green house gases.

```python
# plotting scatter plot for air pollution and forest area
px.scatter(df1,
    x="PM2.5 air pollution, mean annual exposure (micrograms per cubic meter)",
    y="Forest area (% of land area)", 
    color='Country Name',
    title="scatter plot showing total forest VS total air pollution across all the countries")
```

from the above scatter plot it is obserevd that as the forest area increas the amount of air pollution caused is less.

```python
# plotting scatter plot for green house gas and air pollution
px.scatter(df1,
    x="Total greenhouse gas emissions (kt of CO2 equivalent)",
    y="PM2.5 air pollution, mean annual exposure (micrograms per cubic meter)", 
    color='Country Name',
    title="scatter plot showing total green house gas VS total air pollution across all the countries")
```

from the above scatter plot it is obserevd that emission of green house gases and air pollution has linear relationship, as the emission of green house gases increases the air pollution is also increasing.
