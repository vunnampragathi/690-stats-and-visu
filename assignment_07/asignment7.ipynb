```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/wcj365/python-stats-dataviz/master/assignments/assignment_07_world_dev/world_devlopment_indicator.csv")

df.shape
```

```python
df.head()
```

```python
df.sample(5)
```

### You must pick one indicator out of the four: population, life expectancy, GDP per capita, or suiside rate

```python
df["indicator"].unique()             # checking what all the indicators present in the dataset
```

```python
df1 = df.query('indicator == "SH.STA.SUIC.P5" ')    #considering sucide rate indicator

df1
```

```python
df1.tail()
```

### You must pick a few countries of your interest and compare them to see how they differ in the indicator

```python
df["Country Name"].unique()           # checking all the different countries present in the dataset
```

```python
my_countries = ["Greenland", "United Kingdom", "Sri Lanka","Saudi Arabia","United States","Sudan","Philippines"]

df2 = df1[df1["Country Name"].isin(my_countries)]

df2.shape
```

```python
df2.sample(10)
```

```python
df2.tail()
```

### Compare one year (pick the latest year) using bar chart, pie chart, and boxplot

```python
df3 = df2.query('Year == 2019')   # getting 1 year data

df3
```

```python
fig = px.bar(
    df3, 
    x="Country Name", 
    y="value", 
    title="2019 sucide rate of Greenland, United Kingdom, Sri Lanka,Saudi Arabia,United States,Sudan,Philippines ",
    color="Country Name", 
    template="plotly_dark"
)

fig.update_layout(showlegend=False)

fig.show()
```

```python
fig = px.pie(
    df3, 
    names="Country Name", 
    values="value", 
    title=" 2019 sucide rate of Greenland, United Kingdom, Sri Lanka,Saudi Arabia,United States,Sudan,Philippines", 
    color="Country Name", 
    template="plotly_dark"
)

fig.update_layout(showlegend=False)

fig.update_traces(textposition='inside', textinfo='label+percent+value')

fig.show()
```

```python
fig = px.box(df3, x="Country Name", y="value")
fig.show()
```

### Compare all years using line chart

### You must compare all regions to see how they differ in the indicator

```python
df["Region"].unique()    # getting all the regions present in the dataset
```

```python
df2["Region"].unique()
```

```python
df4=df3.groupby('Region')['value'].mean()
df4
```

```python
fig = px.bar(x=df4.index, y=df4)
fig.show()
```

```python
fig = px.pie(df4,names=df4.index,
    values="value", 
    title="2019 sucide rate of all regions",  
    template="plotly_dark"
)
fig.show()
```

### Compare all years using line chart

```python
df5 = df2.groupby(['Region', 'Year']).agg({'value': ['mean']})
df5.columns = ['value_mean']
df5 = df5.reset_index()
print(df5)
```

```python
fig = px.line(
    df5,
    x="Year",
    y="value_mean", 
    title="2019 sucide rate of all the taken regions", 
    color="Region", 
    template="plotly_dark"
)

fig.update_layout(showlegend=True)

fig.show()
```

### Use all countries, and pick year 2020, make a scatter plot with x being GDP per Capita and y being Life Expentancy. This plot will tell you the relationship between wealth and health.

```python
df6 = df.query(' Year == 2020')
df6
```

```python
my_indicators = ["SP.DYN.LE00.IN", "NY.GDP.PCAP.PP.CD"]

df7 = df6[df6["indicator"].isin(my_indicators)]

df7.shape
```

```python
df8 = pd.pivot_table(df7, values = 'value', index=['Country Code', 'Country Name','Year'], columns = 'indicator').reset_index()

df8.head()
```

```python
df8.dropna(inplace=True)

df8.head()
```

```python
fig = px.scatter(
    df8,
    x="NY.GDP.PCAP.PP.CD",
    y="SP.DYN.LE00.IN", 
    title="Wealth vs Health", 
    color="Country Name", 
    template="plotly_dark"
)

fig.update_layout(showlegend=False)

fig.show()
```
