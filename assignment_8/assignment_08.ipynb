**6.Cleanse the data to make the year, population, cost, cost per person numerical type**

```python
def cleanse_year(a_string):
    return int(a_string.replace("*",""))    # replacing * present in the year column to convert year column to int type
```

```python
df["Census Year"] = df["Census Year"].apply(cleanse_year)

df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
def cleanse_population(a_string):
    return int(a_string.replace(",",""))   # replacing , in population column to convert to int type
```

```python
df["Total Population"] = df["Total Population"].apply(cleanse_population)

df.head()
```

```python
def cleanse_cost(a_string):
     a_string=a_string.replace("$","").replace(",","")
     if " Billion" in a_string:
        a_string=int(float(a_string.replace(" Billion",""))*1000000000)
     else:
        a_string=a_string
     return int(a_string)
```

```python
df["Census Cost"] = df["Census Cost"].apply(cleanse_cost)

df.head()
```

```python
def cleanse_cost_per_person(avg_cost):
    x=avg_cost.replace("$","").replace("\n","")
    if "cents" in x:
         y=float(x.replace(" cents",""))/100
    else:
         y=float(x)
    return round(y, 6)
    
```

```python
df["Average Cost Per Person"] = df["Average Cost Per Person"].apply(cleanse_cost_per_person)

df.head()
```
